# Website Scraping Project 

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.scrapethissite.com/pages/simple/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
country = soup.find_all('h3', class_="country-name")
country
```

```python
country_list= []
for tag in country:
        country_list.append(tag.get_text(strip=True))
        
```

```python
len(country_list)
```

```python
columns = soup.find_all('strong')
column_set = []
for column_name in columns:
    column_set.append(column_name.get_text(strip=True))
column_set = [item.replace(':', '').strip() for item in column_set]
```

```python
column_set = set(column_set)
column_set = list(column_set)
column_set
```

```python
capital = soup.find_all(class_="country-capital")
capital_cities = []
for cap in capital:
    capital_cities.append(cap.get_text(strip=True)) 
```

```python
capital_cities
```

```python
population = soup.find_all(class_="country-population")
pop = []
for p in population:
    pop.append(p.get_text(strip=True)) 
```

```python
pop
```

```python
area = soup.find_all(class_="country-area")
area_list = []
for a in area:
    area_list.append(a.get_text(strip=True)) 
```
